## Imports

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
from sklearn import svm, datasets
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
from sklearn.model_selection import GridSearchCV
```

## Loading the database

### Read the csv file and study the database

```python
mushroom_data = pd.read_csv('mushrooms-full-dataset.csv', dtype = str)
mushroom_data.head()
```

```python
mushroom_data['poisonous'].value_counts()
```

- data is balanced

```python
mushroom_data.isnull().sum()
```

- data contains no null values

## Preprocessing

### Defining the target and the inputs

```python
target = mushroom_data['poisonous']
inputs = mushroom_data.drop(['poisonous'], axis = 1)
```

### Creating a training and a testing dataset

```python
x_train, x_test, y_train, y_test = train_test_split(inputs,
                                                    target,
                                                    test_size = 0.2,
                                                   random_state = 365,
                                                   stratify = target)
```

### Checking the result from the stratification

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

### Defining a seperate encoder for the target and the inputs

```python
enc_i = OrdinalEncoder()
enc_t = LabelEncoder()
```

- Label encoding helps with preserving the target labels
- Ordinal encoding transforms categorical data into numerical

### Applying the fit_transform() method on the training data and the transform() method on the test data

```python
x_train_transf = enc_i.fit_transform(x_train)
x_test_transf = enc_i.transform(x_test)

y_train_transf = enc_t.fit_transform(y_train)
y_test_transf = enc_t.transform(y_test)
```

```python
y_train_transf
```

```python
x_train_transf
```

## Rescaling

### Working with SVM's requires specisifc rescaling, inputs need to be in the range [-1,1]

```python
scaling = MinMaxScaler(feature_range= (-1,1)).fit(x_train_transf)
x_train_rescaled = scaling.transform(x_train_transf)
```

## Classification

```python
C = 1.0
svc = svm.SVC(kernel='linear', C = C).fit(x_train_rescaled, y_train_transf)
```

```python
y_pred_test = svc.predict(scaling.transform(x_test_transf))
```

## Model Evaluation

```python
fig, ax = plt.subplots(figsize=(8,5))

cmp = ConfusionMatrixDisplay(
    confusion_matrix(y_test_transf, y_pred_test),
    display_labels = ['Edible', 'Poisonous'],
)

cmp.plot(ax = ax);
```

```python
print(classification_report(y_test_transf, y_pred_test, target_names = ['Edible', 'Poisonous']))
```

- Misclassification rate of the poisonous class is around 7%

## Hyperparameter Tuning with GridSearchCV

### Choose the best kernel and optimal C parameter based on the Cross Validation of the training data

```python
tuned_parameters  =[
    {'kernel':['linear'], 'C':[1,10]},
    {'kernel':['poly'], 'C':[1,10]},
    {'kernel':['rbf'], 'gamma':[1e-3,1e-4], 'C': [1,10]}
]
```

### Gamma controls the radius of the area of the support vector boundary
- Large gamma - Model overfits
- Small gamma - Model underfits

```python
scores = ['precision', 'recall']
```

```python
for score in scores:
    print('# Tuning hyper-parameters for %s' % score)
    print()
    
    clf = GridSearchCV(svm.SVC(), tuned_parameters, scoring = '%s_macro' % score)
    clf.fit(x_train_rescaled, y_train_transf)
    
    print('Best parameters set found on the development set:')
    print()
    print(clf.best_params_)
    print()
    print('Grid scores on development set:')
    print()
    means = clf.cv_results_['mean_test_score']
    stds = clf.cv_results_['std_test_score']
    for mean, std, params in zip(means, stds, clf.cv_results_['params']):
        print('%0.3f (+/-%0.03f) for %r' % (mean, std * 2, params))
    print()
    
    print('Detailed classification report:')
    print()
    print('The model is trained on the full development set.')
    print('The scores are computed on the full evaluation set.')
    print()
    y_true, y_pred  = y_test_transf, clf.predict(scaling.transform(x_test_transf))
    print(classification_report(y_true, y_pred))
    print()
```
